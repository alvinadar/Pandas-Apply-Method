```python
#With this logic i want to apply on the dataframe (Column:Engine_Size_L )
#Step 1 : Define the logic
Engine_size_L = 5

if Engine_size_L >= 2.0:
  print("Big Engine Size.")
else:
  print("Small Engine Size.")
```

```python
#Step 2 : Create the function based on the defined logic

def label_engine_size(engine_size):
  if engine_size >= 2.0:
    return "Big Engine"
  else:
    return "Small Engine"
```

```python
#Use the defined logic and apply on my dataframe (Column:Engine_Size_L )
#apply function will help to broadcast all the function on the Engine_size_column

df['Engine_size_label'] = df['Engine_Size_L'].apply(label_engine_size)
```

#Get age of the car from the data frame


```python
#Age of the car

#Define the logic how to get the age
#Current_Year - Year

#Question , in python how do we get the current year ?

#There is one library in python that can extract all the time data -> We use the datetime

from datetime import datetime

Current_Year = datetime.now().year#Extract only the year from datetime function

def age(year):
  Current_Year = datetime.now().year#Gets the current year
  car_age = Current_Year - year#Gets the age of the car
  return car_age#Returns the car age
```

```python
df['age'] = df['Year'].apply(age)
```

```python
df.head()
```

```python
#Create a new df , where we take the price x sales volume in the df , use apply method .

#Total Sales = Price_USD * Sales_Volume

def total_sales(Price,Sales):
  return Price*Sales

df['Total Sales'] = df[['Price_USD','Sales_Volume']].apply(lambda df:total_sales(df['Price_USD'],df['Sales_Volume']),axis=1)

#Define the manipulation is done on the rows or columns
#Rows -> Axis = 0
#Columns -> Axis = 1

#In the case where we need to involve more than 1 columns we need to use lambda
#Lambda is a simplified version of function
```

#What is a Lambda Function?

		A lambda is a small, anonymous function in Python.
		Defined using the lambda keyword (instead of def).
		They can have any number of inputs but only one expression (the result is returned automatically).

#Syntax

`lambda arguments: expression`

Equivalent to:

`def function_name(argument):
    return expression`

```python

def function_name(argument):
  return #Expression
```

#Simple example :Adding a number

```python
#Using the normal function
x = 5
y = 5

def add(num1,num2):
  return num1+num2

add(5,5)
```

```python
#Use the lambda function

add = lambda num1,num2 : num1+num2

print(add(x,y))
```

#Squaring a number

```python
x1 = 28

def square(num3):
  return num3**2#Power of 2

square(x1)
```

```python
square_lambda = lambda num3:num3**2

square_lambda(x1)
```

#Use lambda for list as well

```python
num_1 = [1,2,3,4,5]

squared = list(map(lambda x : x**2,num_1)
)
print(squared)
```

#Renaming the column in the dataframe

df.rename(columns={'old_column_name','new_column_name'},inplace=True)

```python
df = pd.DataFrame({
    "old": [1, 2, 3],
    "hours": [5, 6, 7]
})
```

```python
df
```

```python
#Rename coumn "old" to "new"

df.rename(columns={'old':'new'})
```

```python
df#Why when we call back the dataframe there is no changes?

#The reason why there is no changes is because there is no reassignment to the original dataframe which is df
```

```python
#There is 2 ways to do reassignment
#Method 1
df = df.rename(columns={'old':'new'})
```

```python
df
```

```python
#Method 2 -> Inplace method
df.rename(columns={'old':'new'},inplace=True)#When you put in the keyword inplace = True -> It will make the changes on the original dataframe
```

#Replacing the values inside the dataframe.

`replace(to_replace,value = None,inplace=True)`

```python
df = pd.DataFrame({
    "Name": ["A", "B", "A", "C"],
    "Grade": ["A", "C", "B", "A"]
})


#Replace the value from A -> B

print(df.replace({'A':'B'},inplace=False))

#Whatever letter that is defined will be replace by the new letter in the dataframe
```

```python
df
```

```python
#Multiple replacement
print(df.replace({'A':'X','B':'Y','C':'Z'},inplace=True))
```
