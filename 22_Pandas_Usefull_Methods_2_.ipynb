#Grouping in dataframe : "Split the dataset into groups based on a column (or multiple columns), then apply an aggregation (like mean, sum, count) to each group."

##Syntax

df.groupby('col')['value'].afffunc()

* df.groupby('col') → Splits data into groups based on the unique values of col.
* ['value'] → Selects the column we want to analyze.
* .mean() (or .sum(), .count(), etc.) → Applies aggregation.

###Common aggregation:

.mean() → average

.sum() → total

.count() → how many records

.min() / .max() → smallest / largest

.agg(['mean', 'count']) → multiple at once

```python
df.groupby(['mental_health_history','social_support'])['stress_level'].mean()
#Give the average stress level for every combination of mental health history and social support.
```

#Pivot Table

##Pivoting reshapes your data or summerizes your data value across two categorical dimension(rows & columns).
##Similar to excel pivot .

`Syntax : df.pivot_table(values='val',index = 'col1',columns = 'col2',aggfunc='mean')`

values='val' → the numeric column you want to summarize.

index='col1' → what goes on the rows.

columns='col2' → what goes on the columns.

aggfunc='mean' → how you summarize (can be mean, sum, count, etc.).

```python
pivot = df.pivot_table(values='stress_level',index='living_conditions',columns='social_support',aggfunc='mean')
```

```python
pivot
```

```python
#Count the student in each group


df.pivot_table(values='stress_level',index='living_conditions',columns='social_support',aggfunc='count')
```

```python
# Maximum stress level observed
df.pivot_table(values='stress_level',index='living_conditions',columns='social_support',aggfunc='max')
```
